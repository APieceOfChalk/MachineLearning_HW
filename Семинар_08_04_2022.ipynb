```python
import networkx as nx
import matplotlib.pyplot as plt
import random
from pyvis.network import Network
```

```python
G = nx.gnm_random_graph(20, 50)

attrs = {}
for i in range(len(G.nodes(data=True))):
    attrs[i] = {"current_load": random.randint(1,100), "max_load": random.randint(1,100)}
nx.set_node_attributes(G, attrs)

plt.figure(3,figsize=(16,16)) 
pos = nx.spring_layout(G)

nx.draw(G, pos)

_labels = nx.get_node_attributes(G,'current_load')
nx.draw_networkx_labels(G, pos, labels = _labels, font_size = 15)


plt.show()
```
